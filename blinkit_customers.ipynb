```python
customers = pd.read_csv('blinkit_customers.csv')
customers_feedback = pd.read_csv('blinkit_customer_feedback.csv')
performance = pd.read_csv('blinkit_marketing_performance.csv')
orders = pd.read_csv('blinkit_orders.csv')
items = pd.read_csv('blinkit_order_items.csv')
products = pd.read_csv('blinkit_products.csv')
```

```python
performance['date'] = pd.to_datetime(performance['date'], errors='coerce')

# Convert multiple date columns in orders.csv
date_columns = ['order_date', 'promised_delivery_time', 'actual_delivery_time']
orders[date_columns] = orders[date_columns].apply(pd.to_datetime, errors='coerce')
```

```python
customers.drop_duplicates(inplace=True)
orders.drop_duplicates(inplace=True)
products.drop_duplicates(inplace=True)
performance.drop_duplicates(inplace=True)
customers_feedback.drop_duplicates(inplace=True)
items.drop_duplicates(inplace=True)
```

```python
customers.columns = [col.strip().lower().replace(' ', '_') for col in customers.columns]
orders.columns = [col.strip().lower().replace(' ', '_') for col in orders.columns]
products.columns = [col.strip().lower().replace(' ', '_') for col in products.columns]
items.columns = [col.strip().lower().replace(' ', '_') for col in items.columns]
customers_feedback.columns = [col.strip().lower().replace(' ', '_') for col in customers_feedback.columns]
performance.columns = [col.strip().lower().replace(' ', '_') for col in performance]
```

```python
for df in [customers, orders, products, items, customers_feedback, performance]:
    for col in df.select_dtypes(include='object'):
        df[col] = df[col].str.strip().str.lower()
```

```python
# Example: Monthly sales trend
orders['order_date'] = pd.to_datetime(orders['order_date'])
monthly_sales = orders.groupby(orders['order_date'].dt.to_period('M'))['order_total'].sum()
monthly_sales.plot(kind='bar')
```

```python
orders_customers = orders.merge(customers, on='customer_id', how='left')
```

```python
orders_customers_feedback = orders_customers.merge(customers_feedback,on=['order_id', 'customer_id'], how='left')
```

```python
orders_items = orders_customers_feedback.merge(items, on='order_id', how='left')
```

```python
master_df = orders_items.merge(products, on='product_id', how='left')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Ensure these columns exist and are correct in your merged DataFrame
# Assumes 'master_df' already contains merged and cleaned data

# Convert date columns
master_df['order_date'] = pd.to_datetime(master_df['order_date'], errors='coerce')
master_df['promised_delivery_time'] = pd.to_datetime(master_df['promised_delivery_time'], errors='coerce')
master_df['actual_delivery_time'] = pd.to_datetime(master_df['actual_delivery_time'], errors='coerce')
master_df['feedback_text'] = master_df['feedback_text'].fillna("")

# Create revenue column
master_df['item_revenue'] = master_df['quantity'] * master_df['unit_price']

# Set up plotting
fig, axs = plt.subplots(3, 2, figsize=(18, 18))

## 1 Top 10 Revenue-Generating Products
try:
    top_revenue = master_df.groupby('product_name')['item_revenue'].sum().sort_values(ascending=False).head(10)
    sns.barplot(x=top_revenue.values, y=top_revenue.index, ax=axs[0, 0])
    axs[0, 0].set_title("Top 10 Revenue-Generating Products")
except Exception as e:
    axs[0, 0].text(0.5, 0.5, "Error in Revenue Plot\n" + str(e), ha='center')

## 2 Most Reviewed Products
try:
    most_reviewed = master_df.groupby('product_name')['feedback_id'].count().sort_values(ascending=False).head(10)
    sns.barplot(x=most_reviewed.values, y=most_reviewed.index, ax=axs[0, 1])
    axs[0, 1].set_title("Most Reviewed Products")
except Exception as e:
    axs[0, 1].text(0.5, 0.5, "Error in Reviews Plot\n" + str(e), ha='center')

## 3 Negative Feedback Word Cloud
try:
    negative_reviews = master_df[
        (master_df['rating'] <= 2) |
        (master_df['sentiment'].str.lower() == 'negative')
    ]
    text = ' '.join(negative_reviews['feedback_text'].dropna().astype(str))
    if len(text.strip()) == 0:
        raise ValueError("No negative feedback to generate word cloud.")
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    axs[1, 0].imshow(wordcloud, interpolation='bilinear')
    axs[1, 0].axis('off')
    axs[1, 0].set_title("Common Words in Negative Feedback")
except Exception as e:
    axs[1, 0].axis('off')
    axs[1, 0].text(0.5, 0.5, "No negative feedback found\n" + str(e), ha='center')

## 4 Monthly Order Trend
try:
    monthly_orders = master_df.dropna(subset=['order_date']).copy()
    monthly_orders['month'] = monthly_orders['order_date'].dt.to_period('M')
    monthly_trend = monthly_orders.groupby('month').size()
    monthly_trend.plot(kind='line', marker='o', ax=axs[1, 1], color='green')
    axs[1, 1].set_title("Monthly Order Trends")
    axs[1, 1].set_xlabel("Month")
    axs[1, 1].set_ylabel("Orders")
except Exception as e:
    axs[1, 1].text(0.5, 0.5, "Error in Monthly Trend Plot\n" + str(e), ha='center')

## 5 Delivery Performance
try:
    delivery_df = master_df.dropna(subset=['promised_delivery_time', 'actual_delivery_time']).copy()
    delivery_df['delayed'] = delivery_df['actual_delivery_time'] > delivery_df['promised_delivery_time']
    delivery_perf = delivery_df['delayed'].value_counts().rename({True: 'Delayed', False: 'On Time'})
    sns.barplot(x=delivery_perf.values, y=delivery_perf.index, ax=axs[2, 0], palette='coolwarm')
    axs[2, 0].set_title("Delivery Performance")
except Exception as e:
    axs[2, 0].text(0.5, 0.5, "Error in Delivery Performance\n" + str(e), ha='center')

## 6 Customer Segments with Highest Revenue
try:
    segment_revenue = master_df.groupby('customer_segment')['item_revenue'].sum().sort_values(ascending=False)
    sns.barplot(x=segment_revenue.values, y=segment_revenue.index, ax=axs[2, 1])
    axs[2, 1].set_title("Customer Segments by Revenue")
except Exception as e:
    axs[2, 1].text(0.5, 0.5, "Error in Segment Revenue Plot\n" + str(e), ha='center')

plt.tight_layout()
plt.show()
```
